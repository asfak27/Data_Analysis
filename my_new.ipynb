```python
import pandas as pd
import re
import requests
def get(link):
    try:
        response = requests.head(link, allow_redirects=True)
        actual_link = response.url
        return actual_link
    except Exception as e:
        print("Error fetching URL:", e)
        return None
def extract_links(text):
    if pd.isna(text):
        return []

    url_pattern = r'https?://\S+'
    links = re.findall(url_pattern, str(text))
    return links

csv3 = pd.read_csv("C:\\jupyter_work\\Main-Dataset.csv", usecols=[5])
column_values = csv3.iloc[:, 0].tolist()

arr1 = []
arr=[]
for text in column_values:
    links = extract_links(text)
    if len(links) == 0:
        arr1.append("NULL")
    else:
        arr1.append(links)
        
csv1 = pd.read_csv("C:\\jupyter_work\\Main-Dataset.csv")
df = pd.DataFrame(csv1)

df["Extract_links"] = arr1
df
```

```python
df.to_csv("C:\\jupyter_work\\Main-Dataset-final.csv")
```
