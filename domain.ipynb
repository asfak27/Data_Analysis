```python
import re

def extract_domains(text):
    # Regular expression pattern to match domains
    domain_pattern = r'(https?://)?([a-zA-Z0-9-]+(?:\.[a-zA-Z]+)+)(?:[/?]|$)'
    
    # Find all matches of the pattern in the text
    matches = re.findall(domain_pattern, text)
    
    # Extract domains from matched groups
    domains = [match[1] for match in matches]
    
    return domains

# Given string
text = "https://davidwalsh.name/git-hook-npm-install-package-json-modified"

# Extract domains
domains = extract_domains(text)
print("Domains:", domains)
```

```python
import re
from collections import Counter

def extract_and_count_domains(texts):
    domain_counter = Counter()
    # Regular expression pattern to match domains
    domain_pattern = r'https?://([^/,]+)'
    
    for text in texts:
        # Find all matches of the pattern in the text
        matches = re.findall(domain_pattern, text)
        
        # Count occurrences of each domain in this text
        domain_counter.update(matches)
    
    return domain_counter


csv3 = pd.read_csv("C:\\Users\\asfak\\OneDrive\\Desktop\\all_in_one.csv", sep='\t',usecols=[3])

df = pd.DataFrame(csv1)
print(df)


csv3 = csv3.applymap(str)

column_values = csv3.iloc[:, 0].tolist() 

# Extract and count domains
domain_counter = extract_and_count_domains(column_values)

# Print the result
print("Domain Counts:")
for domain, count in domain_counter.items():
    print(f"{domain}: {count}")
```

```python
import re
import pandas as pd
from collections import Counter

def extract_and_count_domains(texts):
    domain_counter = Counter()
    # Regular expression pattern to match domains
    domain_pattern = r'https?://([^/,]+)'
    
    for text in texts:
        # Find all matches of the pattern in the text
        matches = re.findall(domain_pattern, text)
        
        # Count occurrences of each domain in this text
        domain_counter.update(matches)
    
    return domain_counter


csv3 = pd.read_csv("C:\\jupyter_work\\Final-Link.csv", usecols=[8], encoding='latin-1')


df = pd.DataFrame(csv3)
#print(df)


csv3 = csv3.applymap(str)

column_values = csv3.iloc[:, 0].tolist() 

# Extract and count domains
domain_counter = extract_and_count_domains(column_values)

# Print the result
print("Domain Counts:")
for domain, count in domain_counter.items():
    print(f"{domain}: {count}")

domain_counts_df = pd.DataFrame(list(domain_counter.items()), columns=['Domain', 'Count'])

# Save domain counts to CSV file
domain_counts_df.to_csv("domain_counts.csv", index=False)    
    
```
